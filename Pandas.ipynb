```python
# imputation ===> to fill the null values or duplicates are treated in such a way that 
#it does not change the statistical meaning of the entire dataset.

# the way the categorical and numerical are treated is different 

# categorical =====> mean
# numerical ========> median
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#for categorical 

df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df["Gender"].mode()[0]
```

```python
# for numerical

df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median)
```

```python
df.info()
```

```python
df["Married"] = df["Married"].fillna(df["Married"].mode()[0])
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])
df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].mode()[0])
```

```python
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].median())
df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].median())
```

```python
#feature engineering:

df["ratio"]=df["ApplicantIncome"]/df["LoanAmount"]
```

```python
# binning
df 
#if person is married, Graduate and NOT se Employed then ===> Stable
#else:
# unstable
```

```python
def bin(x):
    if (x["Married"] == "Yes") & (x["Education"] == "Graduate") & (x["Self_Employed"] == "No"):
        return "Stable"
    else:
        return "Unstable"
```

```python
df["Bin"]=df.apply(bin,axis=1)
```
