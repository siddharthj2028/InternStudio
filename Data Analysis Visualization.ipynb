# LINE PLOTS: 


```python
#create a Dataframe
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a dictionary with sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [2500, 2700, 3000, 3200, 3100, 3300]
}

# Step 2: Convert to DataFrame
df = pd.DataFrame(data)

# Step 3: Line plot
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='dashed', color='red')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales1': [2500, 2700, 3000, 3200, 3100, 3300],
    'Sales2': [3000, 2800, 4000, 3200, 3100, 4000]
}
df = pd.DataFrame(data)
plt.plot(df["Month"], df["Sales1"],marker="o",linestyle="dashed", color="blue")
plt.plot(df["Month"], df["Sales2"],marker="o",linestyle="dashed",color="red")
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.grid(True)
plt.show()
```

# BAR PLOTS


```python
# Step 1: Create a dictionary with sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [2500, 2700, 3000, 3200, 3100, 3300]
}
df = pd.DataFrame(data)

#create a bar plot

plt.bar(df["Month"],df["Sales"])
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
```

# HISTOGRAM 


```python
import numpy as np

#generate random seed
np.random.seed(42)
data = np.random.normal(loc=50,scale=10,size=100)

#create a data Frame
df = pd.DataFrame({"values":data})
df
```

```python
# prepare the histogram
plt.hist(df["values"], bins=20,color="skyblue",edgecolor="black")
plt.xlabel("valeuss")
plt.ylabel("freequency")
plt.title("HIST PLOT")
plt.show()
```

# box plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Math': [78, 85, 90, 72, 88, 95, 60, 76, 82, 99],
    'Science': [84, 89, 94, 90, 92, 85, 78, 88, 91, 87],
    'English': [70, 75, 80, 85, 90, 95, 100, 65, 60, 55]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([df['Math'], df['Science'], df['English']], labels=['Math', 'Science', 'English'])
plt.title("Box Plot of Exam Scores")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

# Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Scores of students in 5 subjects
data = {
    'Math':     [85, 78, 92, 65, 89, 74, 91, 88, 73, 77],
    'Science':  [80, 76, 90, 60, 85, 70, 89, 84, 71, 75],
    'English':  [78, 82, 88, 70, 80, 76, 90, 85, 72, 74],
    'History':  [72, 75, 80, 68, 78, 73, 85, 81, 70, 76],
    'Computer': [90, 85, 95, 78, 92, 84, 96, 91, 80, 83]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Subject Scores")
plt.show()
```
