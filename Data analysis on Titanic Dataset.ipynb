# data Cleaning and Missing Values Handling: 

```python
# if number of entires is N, you have less than 5-10% missing values -- WE CAN ELIMINATE THOSE ROWS 
titanic = titanic[titanic['Embarked'].notna()]
```

```python
null_counts = titanic.isnull().sum()
if null_counts.any():
    print(null_counts[null_counts>0])
```

```python
# if the number id N, you have 30 - 50% missing values -we can replace the rows with mean, median, mode
titanic['Age'].fillna(titanic['Age'].median(), inplace = True)
```

```python
null_counts = titanic.isnull().sum()
if null_counts.any():
    print(null_counts[null_counts>0])
```

```python
#if the number of entries is N, you have above 70% missing values --use some intuition and eliminate the coloumn
#(687/891)*100 = 77.10%
titanic = titanic.drop("Cabin",axis=1)
```

# detecting outliers using data visualization/seaborn

```python
import seaborn as sns 
sns.boxplot(titanic["Survived"])
# every value is B/W 0 - 1 ( suggesting 0 is dead and 1 is survived)
```

```python
sns.boxplot(titanic["Pclass"])
# everything is divided between 1,2,3 and no pointers detected so no outliers
```

```python
sns.boxplot(titanic["Age"])
```

```python
sns.boxplot(titanic["SibSp"])
# outliers are detected greater than 7
```

```python
sns.boxplot(titanic["Fare"])
# outliers are detected 
titanic = titanic[titanic["Fare"]<500]
```

```python
import numpy as np
```

```python
z_scores = np.abs((titanic[["Age","SibSp"]]-titanic[["Age","SibSp"]].mean())/titanic[["Age","SibSp"]].std())
outliers_threshold = 3
outliers = titanic[(z_scores['Age']>outliers_threshold) | (z_scores['SibSp']>outliers_threshold)]
print(outliers)
```

```python
# winsorize the outliers by replacing hem with the 95th percentile value
age_threshold  = titanic["Age"].quantile(0.95)
SibSp_threshold  = titanic["SibSp"].quantile(0.95)

titanic["Age"] = np.where(titanic["Age"]>age_threshold, age_threshold, titanic["Age"])
titanic["SibSp"] = np.where(titanic["SibSp"]>SibSp_threshold, SibSp_threshold, titanic["SibSp"])
```

```python
#drop unneseccary coloumns

titanic.drop(["PassengerId","Name","Ticket"],axis = 1, inplace=True)
```

```python
titanic.drop_duplicates(inplace = True)
titanic.reset_index(inplace = True, drop = True)
titanic
```

```python
#group by

survived = titanic.groupby("Survived").mean(numeric_only = True)
survived
```

```python
#filtering

female_passengers = titanic[titanic.Sex == "female"]
female_passengers

hihg_value_passengers = titanic[titanic.Fare > 35]
hihg_value_passengers
```
